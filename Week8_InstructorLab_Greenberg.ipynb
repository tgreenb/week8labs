### Task 3.2 – Regex select columns ending in 't'  
Use `filter()` with regex to select **columns that end in the letter 't'**. Preview the first five rows.

### Task 3.3 – Sort descending by Team_size_class, human_work, work_per_human  
Use `sort_values()` and set `ascending=False` for all columns. Preview the top 5 rows.  


```python
(df_copy
 .sort_values(by=['Team_size_class', 'human_work', 'work_per_human'], ascending=False)
 .head()
)
```

### Task 3.4 – Human-bot teams with 3+ bots  
Use `query()` to filter for teams where `Team_type == "human-bot"` and `bot_members_count >= 3`.  


```python
(df_copy
 .query('Team_type == "human-bot" and bot_members_count >= 3')
 .head()
)
```

### Task 3.5 – Human teams, Large, gini >= 0.75  
Use `query()` to find teams where:  
- `Team_type == 'human'`  
- `Team_size_class == 'Large'`  
- `human_gini >= 0.75`  
Return the first five rows.


```python
(df_copy
 .query('Team_type == "human" and Team_size_class == "Large" and human_gini >= 0.75')
 .head()
)
```

### Task 3.6 – Count Small or Large Teams

Use `query()` to count how many teams are classified as either:

- `Team_size_class == 'Small'`
- or `Team_size_class == 'Large'`
  

```python
count_3_6 = df_copy.query(
    "Team_size_class == 'Small' or Team_size_class == 'Large'"
).shape[0]

print("Step 3.6 Result:", count_3_6)
```

### Task 3.7 – Count Small/Large teams with low human_gini

Use `query()` to count how many teams are:
- `Team_size_class == 'Small'` or `Team_size_class == 'Large'`
- with `human_gini <= 0.25`

Return just the count.  


```python
count_3_7 = df_copy.query(
    "(Team_size_class == 'Small' or Team_size_class == 'Large') and human_gini <= 0.25"
).shape[0]

print("Step 3.7 Result:", count_3_7)
```

### Task 3.8 – Count Human Teams in Medium Category

Use `query()` to count how many teams are:
- `Team_type == 'human'`
- and `Team_size_class == 'Medium'`

Return just the count.

```python
count_3_8 = df_copy.query(
    "Team_type == 'human' and Team_size_class == 'Medium'"
).shape[0]

print("Step 3.8 Result:", count_3_8)
```

### Task 3.10 – Rename Columns in a New DataFrame

Create a new DataFrame with:
- `human_gini` → rename to `work_inequality`
- `eval_survival_day_median` → rename to `issue_resolution_time`

Use `.rename()` and preview the result
